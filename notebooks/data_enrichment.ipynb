```python
import numpy as np
import pandas as pd
```

```python
starbucks = pd.read_csv('directory.csv')
display(starbucks)
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_central_business_districts'
df_ws = pd.read_html(url)
```

```python
business_centre_africa = df_ws[0]
business_centre_asia = df_ws[1]
business_centre_europe = df_ws[2]
business_centre_north_america = df_ws[3]
business_centre_oceania = df_ws[4]
business_centre_south_america = df_ws[5]

display(business_centre_asia)
```

```python
pdList = [business_centre_africa, business_centre_asia, business_centre_europe, business_centre_north_america, business_centre_oceania, business_centre_south_america]
business_centres_ww = pd.concat(pdList)
```

```python
#business_centres = business_centres_ww.to_csv('business_centres.csv')
```

```python
merge = pd.merge(starbucks,business_centres_ww,on='City',how='left')
```

```python
complete_starbucks_business = merge.to_csv('starbucks_business.csv')
```
